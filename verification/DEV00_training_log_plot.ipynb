```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
log_path = '/glade/work/ksha/CREDIT_runs/wxformer_6h/backup_multi_stage1/training_log.csv'
log_path2 = '/glade/work/ksha/CREDIT_runs/wxformer_6h/training_log.csv'

logs = pd.read_csv(log_path)
logs2 = pd.read_csv(log_path2)

train_mae = np.array(logs['train_mae'])
valid_mae = np.array(logs['valid_mae'])

train_mae2 = np.array(logs2['train_mae'])
valid_mae2 = np.array(logs2['valid_mae'])


train_mae_full = np.concatenate((train_mae, train_mae2), axis=0)
valid_mae_full = np.concatenate((valid_mae, valid_mae2), axis=0)

plt.plot()
L = np.arange(len(train_mae_full))
plt.plot(L, train_mae_full, label='Training MAE (lr=1E-4, cosine annealing)')
plt.plot(L, valid_mae_full, label='Validation MAE')

plt.axvline(x=15, color='r', linestyle='--', label='Epoch 15 (re-start with lr=1E-5 and cosine)')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('Wxformer 6 hourly multi-step training from 06-48 hr\nTraining and Validation MAE')
```
